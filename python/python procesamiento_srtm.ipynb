```python
#| label: python_srtm_gee
#| eval: false
#| echo: true

import ee
import os
import urllib.request
import zipfile
ee.Authenticate()
ee.Initialize(project="ee-pperillav")
srtm = ee.Image("USGS/SRTMGL1_003")
roi_bogota = ee.Geometry.Point([-74.084, 4.638]).buffer(15000).bounds()

# Configuración del directorio de datos actualizado a data/
directorio_salida = "data"
os.makedirs(directorio_salida, exist_ok=True)
ruta_zip = os.path.join(directorio_salida, "mde_bogota_temp.zip")

# Petición a la API para comprimir recorte a 30m
url_descarga = srtm.getDownloadURL({
    'name': 'mde_bogota_srtm', 
    'scale': 30,
    'crs': 'EPSG:4326',
    'region': roi_bogota
})

urllib.request.urlretrieve(url_descarga, ruta_zip)

with zipfile.ZipFile(ruta_zip, 'r') as zip_ref:
    nombre_original_ee = zip_ref.namelist()[0]
    zip_ref.extractall(directorio_salida)

# Renombrado de estandarización en data/
ruta_antigua = os.path.join(directorio_salida, nombre_original_ee)
ruta_nueva = os.path.join(directorio_salida, "mde_bogota_srtm_elevation.tif")
os.rename(ruta_antigua, ruta_nueva)
os.remove(ruta_zip)

print("Descarga nativa SRTM finalizada y estandarizada en la carpeta data/")
```
